## Total Count of Attrition

```python
att_count=(data["Attrition"].value_counts())
att_count
```

## Distribution of Attrition

```python
plt.pie(att_count,labels=['No','Yes'],autopct='%.1f%%', explode=(0.2,0))
plt.legend()
plt.show()
```

```python
data['Attrition'].value_counts().plot(kind='bar', figsize=(12,4), color="red", fontsize=13)

plt.xlabel("Count of people", fontsize=12)
plt.ylabel("Attrition", fontsize=12)
plt.title("Attrition of the employees", fontsize=16)
plt.show()
```

## Age and Attrition

```python
data.hist(column="Age",
         grid=False,
         figsize=(6,4),
         color="blue",
         edgecolor="black",
         bins=12)
plt.xlabel("Age values",fontsize=12)
plt.ylabel("frequency of emp",fontsize=12)
plt.title("Histogram of Age",fontsize=16)
plt.show()
```

## What age group is more likely to leave?

```python
plt.subplots (figsize=(15,4))
sns.countplot (x='Age',  hue='Attrition', data= data, palette = 'colorblind')
plt.show()
```

```python
sns.relplot(x="JobSatisfaction",y="MonthlyIncome",hue="Attrition",kind="line",data=data)
plt.show()
```

```python
tab = pd.crosstab(data["MaritalStatus"],data["Attrition"],normalize='index')*100

tab.plot(kind='bar')
plt.show()
```

## Single people are more likely to leave organization

## WorkLifeBalance

```python
tab = pd.crosstab(data["WorkLifeBalance"],data["Attrition"],normalize='index')*100

tab.plot(kind='bar')
plt.show()
```

People with work life balance of 1 have high attrition rate compared to the population attrition rate of 
Hence poor work life balance increases attrition.

## Years At Company

```python
plt.subplots (figsize=(22,8))
sns.countplot (x='YearsAtCompany',  hue='Attrition', data= data, palette = 'colorblind')
plt.show()
```

Employees who have worked for less number of years tend to leave more.

## Distance From Home

```python
data.groupby(['Age','Attrition'])['DistanceFromHome'].size().unstack().plot(kind='bar', figsize=(12,6),stacked=True)
plt.show()
```
